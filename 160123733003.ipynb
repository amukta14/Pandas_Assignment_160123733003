```python
import pandas as pd
import numpy as np

df = pd.read_csv("Titanic-Dataset.csv")
df.head()
```

```python
#Q.1) 
#Step-1: Create age series
age_series = df['Age']
age_series.head()

#Step-2: Check distribution using descriptive stats
age_series.describe()
```

```python
#Step-3: Create age groups
bins = [0, 12, 18, 35, 60, 100]
labels = ['Child', 'Teen', 'Young Adult', 'Adult', 'Senior']

df['Age_Group'] = pd.cut(df['Age'], bins=bins, labels=labels)
```

```python
#Step-4: Calculate fatality rate
fatality_rate = df.groupby('Age_Group')['Survived'].apply(lambda x: 1 - x.mean())
fatality_rate
```

```python
#Step-5: Find highest fatality group
fatality_rate.idxmax()
```

```python
## Age Distribution and Fatality Analysis

The descriptive statistics show that the average passenger age is around 29.7 years, with most passengers falling between 20 and 38 years, indicating a concentration in the young adult to adult range.

Passengers were grouped into age categories, and fatality rates were calculated for each group.

Key Finding:  
The highest fatality rate was observed in the Senior age group.

Justification:
Children had lower fatality due to priority rescue, while seniors had the highest fatality likely due to reduced mobility and lower evacuation priority.

Conclusion:
Fatality increases with age, with older passengers having lower chances of survival.
```

```python
#Q.5)

# Step-1: Create FamilySize column
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1

# Step-2: Create SurvivalScore using ufuncs
# (normalize Fare to avoid scale issues)
df['Fare_norm'] = df['Fare'] / df['Fare'].max()

# SurvivalScore = weighted combination
df['SurvivalScore'] = (0.7 * df['Fare_norm']) + (0.3 * (df['FamilySize'] / df['FamilySize'].max()))

# Step-3: Create ranked DataFrame
ranked_df = df.sort_values(by='SurvivalScore', ascending=False)

ranked_df[['PassengerId', 'Fare', 'FamilySize', 'SurvivalScore']].head()
```

```python
## Derived Features and Survival Score

A new column FamilySize was created using SibSp + Parch + 1. A SurvivalScore was computed using normalized Fare and FamilySize.
    
Observation:  
Passengers with the highest SurvivalScore have very high fares (≈512.33) and small to moderate family sizes (1–2). Passengers with larger families (size 6) but lower fares have comparatively lower scores.

Conclusion:  
Fare has a stronger influence on SurvivalScore, indicating that higher-paying passengers rank higher, while FamilySize contributes moderately to the score.
```

```python
#Q.8)

# Step-1: Function to categorize age
def categorize_age(age):
    if pd.isna(age):
        return 'Unknown'
    elif age < 18:
        return 'Child'
    elif age < 60:
        return 'Adult'
    else:
        return 'Senior'

# Step-2: Apply function
df['Age_Category'] = df['Age'].apply(categorize_age)

# Step-3: Use map() 
df['Age_Category'] = df['Age_Category'].map({
    'Child': 'Child',
    'Adult': 'Adult',
    'Senior': 'Senior',
    'Unknown': 'Unknown'
})

# Step-4: Create pivot table
pivot_table = pd.pivot_table(
    df,
    values='Survived',
    index='Age_Category'
    columns='Sex',
    aggfunc='mean'
)

pivot_table
```

```python
## Age Categorization and Pivot Table Analysis

Passengers were categorized into Child, Adult, Senior, and Unknown using `apply()`, and the results were stored in the Age_Category column. A pivot table was then created with Age_Category as the index and Sex as the columns to show survival rates.

Observation:
- Female survival rates are much higher than male survival rates in every age category.  
- Senior females have the highest survival rate (1.000), while senior males have a very low survival rate (~0.136).  
- Among males, children have a relatively higher survival rate (~0.397) than adults and seniors.

Conclusion:
The pivot table shows that both age category and gender affect survival, but gender has the stronger influence, with females consistently showing higher survival rates.
```
